```python
import random

class Character:
    def __init__(self, name, level, health, attack, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        get_damage = damage - self.defense
        if get_damage < 0:
            get_damage = 0
        self.health -= get_damage
        if self.health < 0:
            self.health = 0
        print(f"👿{self.name} {get_damage}만큼 얻어 맞았습니다. (남은 체력: {self.health})")

    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f"⚡{self.name} {target.name} 폭행하여 {damage}만큼 아파합니다")
        target.take_damage(damage)

class Player(Character):
    def __init__(self, name):
        super().__init__(name, level = 1, health = 100, attack = 25, defense = 5)
        self.ex = 0

    def gain_experience(self, exp):
        self.ex += exp
        print(f"\n   🎁 {self.name} 경험치 {exp}만큼 얻음 (현재 경험치: {self.ex})")
        self.level_up()

    def level_up(self):
        while self.ex >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.ex -= 50
            print(f"   🎉 {self.name} 레벨업! (레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense})")

class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        super().__init__(name, level, health, attack, defense)

def Battle(player, monster):
    print(f"\n ⚔️  {player.name} vs {monster.name} ⚔️")

    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)

    if player.is_alive():
        print("\n   🏆 전투 승리!!🏆")
        exp_gain = monster.level * 20
        player.gain_experience(exp_gain)
    else:
        print("\n 🔪🔪🔪🔪전투 패배..🔪🔪🔪🔪")


def Main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3, '드래곤': 4, '마동석': 5}
    player_name = input("플레이어 이름을 입력하세요: ")
    player = Player(player_name)

    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)

        Battle(player, monster)  # 전투 시작!

        if not player.is_alive():
            print("\n 💀💀💀💀  게임오버  💀💀💀💀")
            return  # 플레이어가 죽으면 게임 종료

    print(" \n ❤🧡💛💚💙💜🤎  게임 클리어  🤎💜💙💚💛🧡❤ ")

Main()
```
